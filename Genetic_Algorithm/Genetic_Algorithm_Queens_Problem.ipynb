```python
from pickle import POP
import random
import numpy as np;
print("Input Board size : ")
BOARD_SIZE = int(input());
POPULATION_SIZE = 4;

def max_fitness():
    max=0;
    n=BOARD_SIZE;
    i=1;
    while (n-i)!=0:
      max= max + (n-i);
      i+=1;
    return max;

MAX_FITNESS = max_fitness();





#Initial Random Population
def create_initial_population(p):
    queen_placements = []
    for i in range(p):
        placement = []
        for j in range(BOARD_SIZE):
            placement.append(random.randint(0, BOARD_SIZE-1))
        queen_placements.append(placement)
    return queen_placements;



#Non Attacking Pair Fitness
def Calculate_Fitness(queen_placement):
    non_attacking_pairs = 0
    for i in range(len(queen_placement)):
        for j in range(i+1, len(queen_placement)):
            if queen_placement[i] != queen_placement[j] and abs(queen_placement[i]-queen_placement[j]) != j-i:
                non_attacking_pairs += 1
    fitness = non_attacking_pairs ;
    return fitness



def Roulette_Wheel_Selection(population, fitness_values):
    total_fitness = sum(fitness_values)
    selection_value = random.uniform(0, total_fitness)
    fitness_sum = 0
    for i in range(len(population)):
        fitness_sum += fitness_values[i]
        if fitness_sum >= selection_value:
            return population[i]


def CROSSOVER(parent1, parent2):
    crossover_point = random.randint(1, BOARD_SIZE-1)
    offsprings = [];
    offsprings.append(parent1[:crossover_point] + parent2[crossover_point:]);
    offsprings.append(parent2[:crossover_point] + parent1[crossover_point:]);
    return offsprings



def MUTATION(individual):
    #PROBABILITY
    for i in range(len(individual)):
        random_value = random.random();
        if random_value < 1/BOARD_SIZE:
            new_position = random.randint(0, BOARD_SIZE-1)
            individual[i] = new_position
    return individual;



def Evolution_of_population(population, fitness_values):
    mutation_probability = 0.95
    new_population = []
    child=[]

    for i in range(int(POPULATION_SIZE/2)):

        chromosome_1 = Roulette_Wheel_Selection(population, fitness_values)
        chromosome_2 = Roulette_Wheel_Selection(population, fitness_values)
        child = CROSSOVER(chromosome_1, chromosome_2)

        if random.random() < mutation_probability:
            child[0] = MUTATION(child[0]);
            child[1] = MUTATION(child[1]);

        new_population.append(child[0]);
        new_population.append(child[1]);

    return new_population


#Printing Fitness
def Fitness_Pop_Board(population, fitness) :
    for i in range(POPULATION_SIZE):
        print(i+1,"  [ ",end=" ")
        for j in population[i] :
             print(j+1,end=" ");
        print(" ]   : ",fitness[i]);


#Printing Board
def Print_Chess_Board(population):

    Chess =np.full((BOARD_SIZE,BOARD_SIZE),'.');
    i=0;
    h=BOARD_SIZE;
    arr = []
    for x in range(BOARD_SIZE+1) :
      arr.append(h);h-=1;

    for y in population:
        Chess[arr[y+1]][i] = 'Q';
        i+=1;

    for x in range(BOARD_SIZE):
        for y in range(BOARD_SIZE):
            print(Chess[x][y],end="    ")
        print('\n');



#MAIN FUNCTION
def genetic_algorithm():

    iterations=0;
    population = create_initial_population(POPULATION_SIZE)
    fitness= [];



    #Calculating Fitness of Intial Population
    for i in range(POPULATION_SIZE) :
        fitness.append(Calculate_Fitness(population[i]));


    print("Maximum Fitness = ", MAX_FITNESS);


    #Printing Initial Population
    print("Initial Population");
    print("Generation : ",iterations);
    Fitness_Pop_Board(population,fitness);print();
    print("Wait...")



    #Evolution
    while(MAX_FITNESS not in fitness) :
        iterations+=1;
        population=Evolution_of_population(population,fitness);
        fitness.clear();
        for i in range(POPULATION_SIZE) :
           fitness.append(Calculate_Fitness(population[i]));




    #Maxmimum Fitness Population
    print("Maximum Fitness Achieved");
    print("Generation : ",iterations);
    print();

    Fitness_Pop_Board(population,fitness);print();


    #ChessBoard


    for i in range(POPULATION_SIZE):
        if(fitness[i]==MAX_FITNESS):
             print("\n\n<---<*>  THE CHESS BOARD FOR",i+1,"<*>--->")
             Print_Chess_Board(population[i]);
             break;





#Call To Main Function
if BOARD_SIZE > 3 :
  genetic_algorithm();
else :
  print("ERROR CANNOT OPERATE ON 3X3 BOARD ")

```
